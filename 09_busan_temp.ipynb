```python
fName = "busan_temp_20211101152231.csv"
df = pd.read_csv(fName, encoding="cp949", 
skiprows=11, index_col = 2,
names=['지점번호', '지점명','평균기온','평균최고기온','최고기온','최고기온일자',
'평균최저기온','최저기온', '최저기온일자'])
df.head(2)
```
