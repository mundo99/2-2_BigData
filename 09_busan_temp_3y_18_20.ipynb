```python
import matplotlib.pyplot as plt
```

```python
#한글폰트 설정(한글깨짐 방지)
plt.rc('font', family="HYGothic-Medium") #Hancom Gothic
```

```python
from matplotlib import font_manager, rc
plt.rcParams["font.size"] = 14
plt.figure(figsize=(25,8))
plt.plot(df.index, df["최고기온"], c="r")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.plot(df.index, df["최고기온"], c="r", marker="^")
plt.plot(df.index, df["최저기온"], c="b", marker="v")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
