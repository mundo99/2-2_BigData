```python
# value_counts() : 데이터 프레임에 누락된 데이터의 갯수를 정확하게 파악
# dropna = False : NaN 값을 포함한 누락된 데이터 개수
# dropna = True : NaN 값을 제외한 데이터 개수(default)
print(">> 연령 열에 NaN 값이 포함되어 누락된 데이터 개수 계산하여 출력")
df['연령'].value_counts() # dropna=False
```

```python
# 누락 데이터를 열의 평균값으로 치환
# fillna() : 누락된 데이터의 값을 유요한 값으로 대체(inplace = True:원본 변경)
df2 = df.copy() # 원본 df
# 연령 열의 누락된 데이터를 평균값으로 치환하기 전
print(df2['연령'].head())
# df2.fillna(평균값, inplace=True)
```

```python
# 연령 열의 누락된 데이터를 평균값으로 치환하기 후
ageMean = df2['연령'].mean(axis=0) # 열기준
df2.fillna(ageMean, inplace=True)
print(df2['연령'].head())
```

```python
df2 = df.copy()
df2.set_index("이름",inplace=True)
df2.head(3)
```

```python
print(">> 조약돌의 누락된 성별 데이터를 특정 값으로 치환")
df2.loc["조약돌","성별"] = "여"
df2.head(3)
```

```python
# 누락 데이터를 이웃 행의 데이터 값으로 치환 
# 0 앞, 뒤, 주위값
df2 = df.copy()
df2.iloc[:,4:].head()
```

```python
print(">> 누락된 값을 0으로 치환")
df2.fillna(0).iloc[:,4:].head()
```

```python
print(">> 누락된 값을 앞 행의 치환")
df2.fillna(method='ffill').iloc[:,4:].head()
```

```python
print(">> 누락된 값을 앞 행의 치환")
df2.fillna(method='bfill').iloc[:,4:]
```

```python
print(">> 누락된 값을 양쪽 행의 치환")
df2.interpolate().iloc[:,4:]
```

```python
print("각 열에 누락 된 데이터의 개수 확인")
nanDf = df.isnull()
nanDf.head()
```

```python
print("데이터 총 개수 :", len(df2.index))
for col in nanDf.columns:
    # 각 열에 존재하는 NaN 개수 파악
    nanCnt = nanDf[col].value_counts()
    try:
        # NaN 값이 존재하면 개수 출력
        print(col, ":", nanCnt[True])
    except:
        # NaN 값이 존재하지 않으면 0개 출력
        print(col,":",0)
```
