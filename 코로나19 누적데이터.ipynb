```python
import pandas as pd
```

```python
fName = "서울시 코로나19 확진자 현황_utf8.csv"
df = pd.read_csv(fName)
df.head(6)
```

```python
# 데이터 정보 확인
print("데이터의 크기:", df.shape)
print("데이터의 타입:", df.dtypes)
print("데이터의 컬럼명(필드):", df.columns)
print("데이터의 인덱스명:", df.index)

```

```python
# 결측치 데이터 삭제
df.dropna(axis=1, inplace=True)
df.head(3)
```

```python
df.tail(3)
```

```python
df["확진일"]
```

```python
# 데이터 전처리
date = "2021-11-14"
# 년 - 월 - 일
date = date.split("-")
print("date =", date)
print(date[0])
print(date[1])
print(date[2])
```

```python
# 확진일에서 년,월,일 분리하여 저장
df["year"] = df["확진일"].apply(lambda x:x.split("-")[0]) # 년[0]
df["month"] = df["확진일"].apply(lambda x:x.split("-")[1]) # 월[1]
df["day"] = df["확진일"].apply(lambda x:x.split("-")[2]) # 일[3]
df.head()
```

```python
# year, month, day를 object(string) --> int로 변환하는 과정
df["year"] = pd.to_numeric(df["year"])
df["month"] = pd.to_numeric(df["month"])
df["day"] = pd.to_numeric(df["day"])
df.info()
```

```python
df21 = df[df["year"] == 2021]
df21.tail(3)
```

```python
df20 = df[df["year"] == 2020]
df20.tail(3)
```

```python
# 20년 월별 확진자 수 
df20M = df20["month"].value_counts().sort_index()
df20M
```

```python
# 21년 월별 확진자 수 
df21M = df21["month"].value_counts().sort_index()
df21M
```

```python
# x축 데이터 만들기 1월~12월
monthName = []
for i in range(1,13): # 1~12
    monthName.append(f"{i}월")
print(monthName)
```

```python
# 20년 월별 확진자 수 
df21M = df21["month"].value_counts().sort_index()
df21M
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font',family="HYGothic-Medium")
```

```python
plt.figure(figsize=(10,4))
plt.plot(df21.index,df21M)
plt.xticks(range(1,13),monthName)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(df21.index,df21M, label="21년도 확진자 수")
plt.bar(df20.index,df20M, label="20년도 확진자수")
plt.xticks(range(1,13),monthName)
plt.legend()
plt.show()
```
