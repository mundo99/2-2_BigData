```python
import pandas as pd
```

```python
# 데이터프레임 생성(딕셔너리)
bio_data = {'성명':['이겨레','조약돌'],
            '나이':[25,19], 
            '성별':['남','여'],
            '키':[185,163],
            '체중':[89,52]}
df = pd.DataFrame(bio_data)
df
```
