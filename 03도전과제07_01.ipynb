4. iloc -> 부산지점 행 데이터 출력

```python
# 1. 데이터 프레임 생성 및 출력
import pandas as pd
df = pd.DataFrame({"컴퓨터":[869, 578, 321],
                    "모니터":[374, 286, 148],
                    "노트북":[569, 427, 335],
                    "스마트폰":[197, 89, 66]})
df
```

```python
# 2. 인텍스 설정
df.index = ["서울지점","부산지점","제주지점"]
df
```

```python
# 3. loc -> 서울지점 행 데이터 출력
df.loc["서울지점"]
```

```python
# 4. iloc -> 부산지점 행 데이터 추력
df.iloc[1]
```
