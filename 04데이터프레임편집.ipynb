```python
import pandas as pd
```

```python
bio_data = {'성명':['이겨레','조약돌'],
            '나이':[25,19], '성별':['남','여'],
            '키':[185,163], '체중':[89,53]}
df = pd.DataFrame(bio_data)
df
```

```python
# 열 추가 (like Dict)
df['BMI'] = 0
df
```

```python
df['BMI'] = [18.6, 22.0] 
df
```

```python
# 행 추가
# 1. 모든 원소에 같은 데이터의 값을 추가하여 새로운 행 추가
df.loc[2] = 0
df
```

```python
# 2. 열의 개수만큼 데이터의 값을 각각 지정하여 새로운 행 추가
df.loc[3] = ['강산에',27,'남',177,63,20.9]
df
```

```python
# 3. 기존 행을 복사하여 새로운 행 추가
df.loc['홍길동'] = df.loc[3]
df
```
