```python
# 07중간고사대비실습2
import pandas as pd
```

```python
fName="health_data.csv"
df = pd.read_csv(fName) # encoding="cp949"
df
```

```python
# 6.성명으로 열 추가하기 (추가내용: "이순신","선덕여왕","명성황후")
df["성명"] = ["이순신","선덕여왕","명성황후"]
df
```

```python
# 7.코드번호 열 삭제하기(axis,inplace 옵션 사용하기)
df.drop("코드번호", axis=1, inplace=True)
df
```

```python
# 8.성명을 인덱스로 지정하기
# (원본 수정되도록 옵션 inplace 넣기)
df.set_index("성명",inplace=True)
df
```

```python
# 9. 선덕여왕의 BMI를 20.00 으로 수정하기
df.loc["선덕여왕","BMI"] = 20.00
df
```

```python
# 10. '성별' 열의 고유값 개수 확인하기
df["성별"].value_counts()
```
