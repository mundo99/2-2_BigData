```python
# 데이터 프레임 생성 p223
import pandas as pd 


bio_data = {'성명':['이겨레','조약돌'], 
            '나이':[25,19],
            '성별':['남','여'],
            '키':[185,163],
            '체중':[89,56]
            }
df = pd.DataFrame(bio_data)
df
```
