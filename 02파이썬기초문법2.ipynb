## 딕셔너리 자료형

```python
# 딕셔너리 = {key:value, key:value}
dictInfo = {"name":"문도연","likeNum":2,"height":171.5}
print(dictInfo)
```

```python

# 오직 키값으로만 출력이 가능
print(dictInfo["height"])
```

```python
# 없으면 추가
dictInfo["hobby"] = ["독서","요리","만들기"]
print(dictInfo)
```

```python
# 있으면수정
dictInfo["likeNum"] = 7
print(dictInfo)
```

```python
# 삭제 del dictInfo["height"]
dictInfo.pop("height")
print(dictInfo)
```

```python
dictInfo = {"name":"문도연","likeNum":2,"height":171.5}
dictInfo["hobby"] = ["독서","요리","만들기"]
print("name",dictInfo["name"])
print("likeNum",dictInfo["likeNum"])
print("height",dictInfo["height"])
print("hobby",dictInfo["hobby"])
```

```python
for item in dictInfo.items():
    print(item, type(item)) # 리스트와 동일한 tuple type(읽기전용)
    
```

```python
for k,v in dictInfo.items():
    print(k,v,sep=" = ")
    
```

```python
for item  in dictInfo["hobby"]:
    print(item)
```
