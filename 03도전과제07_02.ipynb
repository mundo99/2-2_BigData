```python
# 1. 데이터 프레임 생성 및 출력
import pandas as pd
df = pd.DataFrame({"컴퓨터":[869, 578, 321],
                    "모니터":[374, 286, 148],
                    "노트북":[569, 427, 335],
                    "스마트폰":[197, 89, 66]},
                    index = ["서울지점","부산지점","제주지점"])
df
```

```python
# 2. 도트(.)를 사용하여 지점명에 대한 1개의 열 데이터(모니터)만 선택하여 출력
df.모니터
```

```python
# 3. 대괄호([])를 사용하여 지점명과 노트북에 대한 2개의 열 데이터(노트북,스마트폰)만 선택하여 출력
df[["노트북","스마트폰"]]
```

```python
# 4. 선택된 데이터의 타입을 출력하기 우해 type()메소드를 사용하기
type(df["컴퓨터"])
df["컴퓨터"]
```
