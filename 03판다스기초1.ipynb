```python
# 데이터프레임: 인덱스(행), 칼럼(열), 2차원 데이터
# 데이터프레임 생성
# 변수명 = 객체명.DataFrame([행들],[열들]],index=[행이름들], columns=[열이름들])

import pandas as pd
df = pd.DataFrame([['환자코드','연령','성별','거주지'],
                       ['A00301',23,'남','서울특별시'],
                       ['D00908',53,'여','제주특별자치도']])
# print(df)
df
                     
```

```python
df = pd.DataFrame([[23,'남','서울특별시'],
                   [53,'여','제주특별자치도']],
                   index=["A00301","D00908"], columns=['연령','성별','거주지'])
# print(df)
df
                     
```

```python
df = pd.DataFrame([[23,'남','서울특별시'],
                   [53,'여','제주특별자치도']])
df.index=["A00301","D00908"]
df.columns=['연령','성별','거주지']
df
  
```

```python
# 인덱스 이름 변경
df.rename(index={"A00301":"환자1","D00908":"환자2"}, inplace=True)
df
```

```python
# 칼럼 이름 변경
df.rename(columns={"연령":"age","성병":"gender","거주지":"addr"}, inplace=True)
df
```

이겨레 25   남  185 89
조약돌 19   여  163 53


```python
df = pd.DataFrame([[25,'남',185,89],
                   [19,'여',163,53]],
                   index=["이겨레","조약돌"], columns=['나이','성별','키','체중'])
# print(df)
df
```

```python
# 행 데이터 선택
# loc[name], iloc[idx]
print(df.loc["이겨레"])
print(df.iloc[0])
```

```python
print(df.iloc[0:2])
```
